```python
# 1.

import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset


# Read the dataset
car_data = pd.read_csv(r"D:\python-venv\myenv\Include\car_dataset1.csv")

# Extract relevant columns for the first 5 models from each car
car_names = []
highway_mpg = []
city_mpg = []

for make in car_data['Make'].unique()[:5]:
    car_subset = car_data[car_data['Make'] == make].head(1)
    car_names.append(car_subset['Make'].values[0])
    highway_mpg.append(car_subset['highway MPG'].values[0])
    city_mpg.append(car_subset['city mpg'].values[0])

# Create a figure and axis object
fig, ax = plt.subplots()

# Define the width of each bar
bar_width = 0.35

# Create bars for highway MPG
bar1 = ax.bar(range(len(car_names)), highway_mpg, bar_width, label='Highway MPG', color='brown')

# Create bars for city MPG
bar2 = ax.bar([x + bar_width for x in range(len(car_names))], city_mpg, bar_width, label='City MPG', color='y')

# Set labels and title
ax.set_xlabel('Cars')
ax.set_ylabel('MPG')
ax.set_title('Highway MPG and City MPG Comparison for Selected Models')
ax.set_xticks([x + bar_width / 2 for x in range(len(car_names))])
ax.set_xticklabels(car_names, rotation=45)  # Rotate x-axis labels for better readability
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()

```

```python
# 2.

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
car_data = pd.read_csv(r"D:\python-venv\myenv\Include\car_dataset.csv")



# Assuming the dataset has columns like 'car_name' and 'popularity', adjust accordingly
# Group by car_name and calculate the total popularity for each car
popularity_by_car = car_data.groupby('Make')['Popularity'].sum().reset_index()

# Calculate the total popularity across all cars
total_popularity = popularity_by_car['Popularity'].sum()

# Calculate the popularity percentage for each car
popularity_by_car['popularity_percentage'] = popularity_by_car['Popularity'] / total_popularity * 100

# Sort the dataframe by popularity in descending order and select the top 10 rows
top_10_popular_cars = popularity_by_car.sort_values(by='Popularity', ascending=False).head(10)

# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_10_popular_cars['popularity_percentage'], labels=top_10_popular_cars['Make'], autopct='%1.1f%%', startangle=140)
plt.title('Popularity of Top 10 Cars')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```

```python
# 10.

import matplotlib.pyplot as plt
import pandas as pd

# Load the car dataset
car_data = pd.read_csv(r"D:\python-venv\myenv\Include\car_dataset.csv")

# Filter cars where Engine Cylinders is greater than 10 and either rear or front or all wheel drive
filtered_front_wheel_drive = car_data[(car_data['Engine Cylinders'] > 10) & 
                                      (car_data['Driven_Wheels'] == 'front wheel drive')]

filtered_rear_wheel_drive = car_data[(car_data['Engine Cylinders'] > 10) & 
                                     (car_data['Driven_Wheels'] == 'rear wheel drive')]

filtered_all_wheel_drive = car_data[(car_data['Engine Cylinders'] > 10) & 
                                    (car_data['Driven_Wheels'] == 'all wheel drive')]

# Count the number of cars for each category
counts_front_wheel_drive = filtered_front_wheel_drive['Make'].value_counts()
counts_rear_wheel_drive = filtered_rear_wheel_drive['Make'].value_counts()
counts_all_wheel_drive = filtered_all_wheel_drive['Make'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))

# Plotting bars for front wheel drive cars
plt.bar(counts_front_wheel_drive.index, counts_front_wheel_drive.values, color='pink', label='Front Wheel Drive', edgecolor='black')

# Plotting bars for rear wheel drive cars
plt.bar(counts_rear_wheel_drive.index, counts_rear_wheel_drive.values, color='yellow', label='Rear Wheel Drive', edgecolor='black')

# Plotting bars for all wheel drive cars
plt.bar(counts_all_wheel_drive.index, counts_all_wheel_drive.values, color='green', label='All Wheel Drive', edgecolor='black')

plt.title('Cars with Engine Cylinders > 10 and either rear or front or all wheel drive')
plt.xlabel('Car Make')
plt.ylabel('Count')

plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()




```

```python
# 12.

import pandas as pd
import matplotlib.pyplot as plt

# Load the car dataset
car_data = pd.read_csv(r"D:\python-venv\myenv\Include\car_dataset.csv")

# Function to create a pie chart for a specific year
def plot_pie_chart_for_year(year):
    # Filter data for the specified year
    filtered_data = car_data[car_data['Year'] == year]
    
    # Group by Market Category and count occurrences
    grouped_data = filtered_data.groupby('Market Category').size().reset_index(name='count')
    
    # Filter for only Luxury and High-Performance categories
    luxury_hp_data = grouped_data[grouped_data['Market Category'].isin(['Luxury', 'High-Performance'])]
    
    # Plotting pie chart
    plt.figure(figsize=(6, 6))
    labels = luxury_hp_data['Market Category']
    sizes = luxury_hp_data['count']
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title(f'Year {year}')
    plt.show()

# Years to analyze
years_of_interest = [2005, 2007]

# Plotting pie charts for individual years
for year in years_of_interest:
    plot_pie_chart_for_year(year)


plt.show()


```

```python
# 13.


import pandas as pd
import matplotlib.pyplot as plt
import random

# Load the car dataset
car_data = pd.read_csv(r"D:\python-venv\myenv\Include\car_dataset.csv")

# Select two random years
random_years = random.sample(range(car_data['Year'].min(), car_data['Year'].max() + 1), 2)

# Filter the dataset for the selected years
filtered_data = car_data[car_data['Year'].isin(random_years)]

# Group the filtered data by 'Year' and 'Make' to count the number of car launches for each make in each year
grouped_data = filtered_data.groupby(['Year', 'Make']).size().unstack(fill_value=0)

# Select only the top five car makes with the highest number of launches
top_5_car_makes = grouped_data.sum().nlargest(5).index
grouped_data = grouped_data[top_5_car_makes]

# Plotting a stacked area chart
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='area', stacked=True)

# Adding labels and title
plt.title('Number of Car Launches by Make for Randomly Selected Years (Top 5 Car Makes)')
plt.xlabel('Year')
plt.ylabel('Number of Car Launches')

plt.legend(title='Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load the car dataset
car_data = pd.read_csv("car_dataset.csv")

# Select two random years
random_years = random.sample(range(car_data['Year'].min(), car_data['Year'].max() + 2), 4)

# Filter the dataset for the selected years
filtered_data = car_data[car_data['Year'].isin(random_years)]

# Find the car with the highest MSRP for each year
max_msrp_cars = filtered_data.loc[filtered_data.groupby('Year')['MSRP'].idxmax()]

# Create a dictionary to store the total MSRP for each year
msrp_by_year = {}
for year in random_years:
    msrp_by_year[year] = max_msrp_cars[max_msrp_cars['Year'] == year]['MSRP'].sum()

# Plotting a pie chart
plt.figure(figsize=(10, 6))

# Plot the pie chart with total MSRP for each year
plt.pie(msrp_by_year.values(), labels=msrp_by_year.keys(), autopct='%1.1f%%', startangle=140)

# Adding labels and title
plt.title('Total MSRP of Highest Cars for Two Randomly Selected Years')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()

```
